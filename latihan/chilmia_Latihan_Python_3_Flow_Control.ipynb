1. Buatlah program untuk menentukan status berat badan berdasarkan Body Mass Index (BMI)!

Rumus BMI: `berat badan / tinggi badan^2`

<br>

Status berat badan berdasarkan BMI:

bmi.png

[Sumber Gambar](https://www.gendhismanis.id/kalkulatorbmi.html).

<br>

Contoh output program:

soal flowcontrol 3.PNG

```python
berat=float(input('berat badan (kg) :'))
tinggi=float(input('tinggi badan (cm) :'))
tinggi= tinggi/100
BMI = berat/tinggi**2


if BMI<=18.5:
    print(f'{BMI} Kurang berat badan')
elif 18.5<=BMI<=22.9:
    print(f'{BMI} normal')
elif 23<=BMI<=24.9:
    print(f'{BMI} kelebihan berat') 
elif 25<=BMI<=29.9:
    print(f'{BMI} Obesitas Tingkat 1')
elif BMI>=30:
    print(f'{BMI} Obesitas tingkat 2')
```

2. Buatlah sebuah program yang akan terus looping sampai user menginput angka nol! Setelah user menginput angka nol, program kemudian akan menghitung rata-rata seluruh angka yang telah diinput oleh user sebelumnya. Contoh output program:

soal flowcontrol 1.PNG


```python
import numpy as np
x = []
while True:
  data = int(input('Silahkan input angka = '))
  if data == 0:
    print('Rata-rata semua angka yang telah diinput sebelumnya adalah ', np.mean(x))
    break
  else:
    x.append(data)
```
