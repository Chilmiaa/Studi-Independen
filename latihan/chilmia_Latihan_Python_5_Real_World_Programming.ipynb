2. Buat sebuah program OOP dengan ketentuan:

  - berbeda dari contoh program yang ada di materi (ppt dan colab)
  - memiliki 1 parent class
  - memiliki setidaknya 2 child class
  - menerapkan konsep dasar OOP (kecuali abstraction)
  - dilengkapi dengan docstring, contoh docstring bisa dilihat dalam program hero.py

  Setelah program selesai, buatlah:
  - setidaknya 2 instance untuk setiap child class
  - Class Diagram-nya

```python
class penjualan:
  ''' OOP ini untuk menginformasikan sistem jual beli barang secara online
  '''

  nama = ''
  nama_barang = ''
  order = ''

  def pembayaran(self):
    print('pembeli melakukan pembayaran')
  def pengiriman(self):
    print('penjual mengirim barang')
  def penilaian(self):
    print('pembeli memberikan penilaian')

class pembeli(penjualan):
  def __init__(self, nama, nama_barang, order):
    self.nama = nama
    self.nama_barang = nama_barang
    self.order = order

  def pembayaran(self):
    if self.order == "ya":
      checkout = input('apakah pembeli melakukan pemesanan?(y/n)')
      if checkout == 'y':
        print(f'{self.nama} memesan barang')
      elif checkout == 'n':
        print(f'{nama} tidak jadi memesan barang')
    else :
      print('pembeli tidak order')

  def penilaian(self):
    barang_diterima = input('apakah barang sudah diterima dengan baik dan benar?(y/n)')
    if barang_diterima == 'y':
      pesanan_diterima = input (f'silahkan ketik (ya) untuk pelepasan dana ke penjual atau dana masuk ke penjual secara otomatis setelah 3 hari :')
      if pesanan_diterima == 'ya':
        print ('Terimakasih, dana akan diteruskan ke penjual')
      penilaian = input(f' apakah anda ingin memberikan penilaian?(y/n)')
      if penilaian == 'y':
        print(input('silahkan masukkan penilaian :'))
    elif barang_diterima == 'n':
      print(f'silahkan hubungi penjual atau anda bisa melakukan ajukan pengembalian barang')

class penjual(penjualan):
  def __init__(self, nama, nama_barang):
    self.nama = nama
    self.nama_barang = nama_barang

  def pengiriman(self):
    payment = input('apakah pembeli melakukan pembayaran?(y/n)')
    if payment == 'y':
      print(f'penjual melakukan pengiriman barang')
    elif payment == 'n':
      print(f'penjual tidak memproses pemesanan')
```

```python
chilmia = pembeli('chilmia', 'paket1', 'ya')
chilmia.pembayaran()
```

```python
sofia = penjual('sofia', 'paket1')
sofia.pengiriman()
```
