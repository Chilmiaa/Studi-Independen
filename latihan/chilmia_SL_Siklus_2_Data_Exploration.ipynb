Visualisasi Data Menggunakan Library Matplotlib

1. Bar Graph

```python
# Library Matplotlib
import matplotlib.pyplot as plt
```

```python
city_count  = dataset[' "City"'].value_counts()
city_count = city_count[:10,]
plt.figure(figsize=(20,5))
plt.bar(city_count.index, city_count.values, alpha=0.8)
plt.title('Perbandingan jumlah setiap Kota')
plt.ylabel('Count', fontsize=12)
plt.xlabel('City', fontsize=12)
plt.show()
```

```python
kota_SF  = dataset[dataset[' "City"']==' "Springfield"']['LatD'].mean()
kota_ST  = dataset[dataset[' "City"']==' "Schenectady"']['LatD'].mean()

kota = [kota_SF, kota_ST]
labels = [' "Springfield"',
          ' "Schenectady"']

plt.figure(figsize=(20, 5))
plt.bar(labels, kota)
plt.title('Perbandingan rata-rata LatD setiap City')
plt.show()
```

Visualisasi Data Menggunakan Library Seaborn

```python
# Library Seaborn
import seaborn as sns
```

1. Line Chart

```python
sns.lineplot(data=dataset[[' "LatM"', ' "LonD"']])
```

2. Scatter Plot

```python
sns.relplot(x=' "LonM"', y='LatD', data=dataset)
```

3. Histogram

```python
sns.displot(data=dataset, x='LatD', hue=' "City"', kde=True)
```

4. Box Plot

```python
sns.boxplot(data=dataset, x='LatD')
```

5. Cat Plot (Category Plot)

```python
sns.catplot(
    data=dataset, x=' "State"', y='LatD',
    height=10, aspect=2.5,
)
```

6. Heatmap (Korelasi)

```python
corrmat = dataset.corr()
plt.figure(figsize=(10, 7))
sns.heatmap(corrmat, annot=True)
```

Warna merepresentasikan keeratan hubungan:
- bisa diliat dari nilai hubungan/korelasinya.
- bisa dilihat dari penggaris di sebelah heatmap.
