## 1. Program Penghitung Upah Karyawan

Program harus mengimplementasikan konsep function yang memanggil function lainnya (*divide and conquer*)! Jangan lupa tambahkan dokumentasi di dalam function menggunakan docstring!

Lalu, hitunglah upah bersih karyawan berikut ini menggunakan program yang telah dibuat:

```python
def Upah_Bersih(nama, gaji_pokok, tunjangan, upah_lembur, overtime):
  print('Nama :', nama)
  tunjangan = gaji_pokok* (tunjangan/100)
  upah_overtime = upah_lembur * overtime
  upah_kotor = gaji_pokok + tunjangan + upah_overtime
  potongan = 0.02 * upah_kotor
  print('Upah_Bersih:', upah_kotor - potongan)

print('List Gaji karyawan')
Upah_Bersih('Anna', 7500000, 10, 150000, 10)
Upah_Bersih('Dimas', 5000000, 5, 100000, 15)
Upah_Bersih('Sriyono', 15000000, 15, 0, 3)
Upah_Bersih('Oky', 7500000, 10, 150000, 8)
Upah_Bersih('Wati', 15000000, 15, 0, 6)
Upah_Bersih('Yayan', 5000000, 5, 100000, 17)
```

## 2. Lambda

Buatlah sebuah list yang berisi elemen dengan kelipatan tanggal lahir kalian, dari 1 sampai dengan 30! List harus dibuat menggunakan lambda!

kelipatan26.PNG


**Note**: Jika dikerjakan secara berkelompok, maka elemen list adalah kelipatan rata-rata tanggal lahir anggota kelompok.

```python
TL = [(lambda x:x*3)(x) for x in range (1,31)]
print(TL)
```
